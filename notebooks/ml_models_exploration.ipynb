```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn import svm
from sklearn.neural_network import MLPRegressor, MLPClassifier

import warnings
warnings.filterwarnings(action='ignore')
```

```python
data = pd.read_csv("fire_data_turkey.csv")
data
```

```python
normalized_data_df = data.copy()
```

```python
data.info()
```

```python
normalized_data_df = normalized_data_df[normalized_data_df['PrecipitationCm'].notna()]
normalized_data_df.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
 
le = LabelEncoder()
label = le.fit_transform(normalized_data_df['Weather type'])
normalized_data_df.drop("Weather type", axis=1, inplace=True)
normalized_data_df["Weather type"] = label

label = le.fit_transform(normalized_data_df['City'])
normalized_data_df.drop("City", axis=1, inplace=True)
normalized_data_df["City"] = label
```

```python
def normalizer(values):
  values = values.reshape((len(values), 1))
  scaler = MinMaxScaler(feature_range=(0, 1))
  scaler = scaler.fit(values)
  normalized = scaler.transform(values)

  return normalized
```

```python
normalized_data_df['Day'] = normalizer(normalized_data_df.Day.values)
normalized_data_df['Month'] = normalizer(normalized_data_df.Month.values)
normalized_data_df['Year'] = normalizer(normalized_data_df.Year.values)
normalized_data_df['AvgDWS'] = normalizer(normalized_data_df.AvgDWS.values)
normalized_data_df['Visibility'] = normalizer(normalized_data_df.Visibility.values)
normalized_data_df['MaxWG'] = normalizer(normalized_data_df.MaxWG.values)
normalized_data_df['MaxSusWS'] = normalizer(normalized_data_df.MaxSusWS.values)
normalized_data_df['PrecipitationCm '] = normalizer(normalized_data_df.PrecipitationCm .values)
normalized_data_df['TempAvgCel'] = normalizer(normalized_data_df.TempAvgCel.values)
normalized_data_df['TempMaxCel'] = normalizer(normalized_data_df.TempMaxCel.values)
normalized_data_df['TempMinCel'] = normalizer(normalized_data_df.TempMinCel.values)
normalized_data_df['DewPntCel'] = normalizer(normalized_data_df.DewPntCel.values)
```

```python
normalized_data_df
```

```python
del normalized_data_df['Date'], normalized_data_df['MaxSusWS'], normalized_data_df['MaxWG'], normalized_data_df['Weather type']
```

```python
X = normalized_data_df.drop('Fire', axis=1)
y = normalized_data_df['Fire']
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, shuffle=True, random_state=1)
```

```python
lin_reg_model = LinearRegression()
lin_reg_model.fit(X_train, y_train)

print("Linear Regression R^2: {:.5f}".format(lin_reg_model.score(X_test, y_test)))
```

```python
lin_clf_model = LogisticRegression()
lin_clf_model.fit(X_train, y_train)

print("Linear Classification Accuracy: {:.2f}%".format(lin_clf_model.score(X_test, y_test) * 100))
```

```python
nn_clf_model = MLPClassifier(hidden_layer_sizes=(16,16))
nn_clf_model.fit(X_train, y_train)

print("NN Classification Accuracy: {:.2f}%".format(nn_clf_model.score(X_test, y_test) * 100))
```

```python
from sklearn.metrics import classification_report
```

```python
svm_model = svm.LinearSVC()
svm_model.fit(X_train, y_train)
print("SVM Accuracy: {:.2f}%".format(svm_model.score(X_test, y_test) * 100))
```

```python
svm_model_acc = svm_model.predict(X_test)
print(classification_report(y_test, svm_model_acc))
```
