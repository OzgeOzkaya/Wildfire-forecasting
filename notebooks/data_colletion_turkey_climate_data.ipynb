```python
parameters = ['Temperature avarage, Fahrenheit / Celsius', 'Temperature maximum, Fahrenheit / Celsius', 'Temperature minimum, Fahrenheit / Celsius', 'Precipitation in / cm', 'Average daily wind speed (miles/hour)',
              'Visibility in miles', 'Dew point, Fahrenheit / Celsius', 'Maximum sustained wind speed (miles/hour)', 'Maximum wind gust reported for the day', 'Sea level pressure in mBARs', 
              'Station pressure in mBARs', 'Snow depth in inches', 'Weather type']
```

```python
def update_dict(records, key_name, update_element):
  # rec = records.get(key_name)
  records[key_name].append(str(update_element))
```

```python
records = {'City': [], 'Year': [], 'Month': [], 'Day': [], 'Temperature avarage, Fahrenheit / Celsius': [], 'Temperature maximum, Fahrenheit / Celsius': [], 'Temperature minimum, Fahrenheit / Celsius': [], 
           'Precipitation in / cm': [], 'Average daily wind speed (miles/hour)': [], 'Visibility in miles': [], 'Dew point, Fahrenheit / Celsius': [], 'Maximum sustained wind speed (miles/hour)': [], 
           'Maximum wind gust reported for the day': [], 'Sea level pressure in mBARs': [], 'Station pressure in mBARs': [], 'Snow depth in inches': [], 'Weather type': []}
```

```python
for city in cities:
  print(city)
  for year in years:
    for month in months:  
      url = "https://geographic.org/global_weather/weather_data_2.php?month={0}&year={1}&id=173520-99999&path=weather_stations/171880_267589/173520-99999.txt&name={2}&country=Turkey".format(month, year, city)

      page = requests.get(url)
      text = page.text
      soup = BeautifulSoup(text)

      for hit in soup.findAll(attrs={'id' : 'table'}):
        hit
      
      t = str(hit).split("<tr>")
      for x in range(len(t)):
        if x < 2:
          continue
        update_dict(records, 'City', city)
        update_dict(records, 'Year', year)
        update_dict(records, 'Month', month)
        # city_name.append(city)
        # year_row.append(year)
        # month_row.append(month)
        row = BeautifulSoup(t[x])
        day_number = row.find('strong').getText()
        # day.append(day_number)
        update_dict(records, 'Day', day_number)

      for x in range(len(t)):
        if x < 2:
          continue
        for param in parameters:
          row = BeautifulSoup(t[x])
          for test in row.findAll(attrs={'title' : param}):
            update_dict(records, param, test.text)
            # avarage_temp.append(test.text)

  df = pd.DataFrame(list(zip(records['City'], records['Year'], records['Month'], records['Day'], records['Temperature avarage, Fahrenheit / Celsius'], records['Temperature maximum, Fahrenheit / Celsius'], 
                           records['Temperature minimum, Fahrenheit / Celsius'], records['Precipitation in / cm'], records['Average daily wind speed (miles/hour)'], records['Visibility in miles'], 
                           records['Dew point, Fahrenheit / Celsius'], records['Maximum sustained wind speed (miles/hour)'], records['Maximum wind gust reported for the day'], records['Sea level pressure in mBARs'], 
                           records['Station pressure in mBARs'], records['Snow depth in inches'], records['Weather type'])), columns=list(records.keys()))
  df.to_csv('/content/drive/MyDrive/wildfire/turkey_daily_climate_data.csv')

      
```
